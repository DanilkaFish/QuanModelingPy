```python
from itertools import combinations
from igraph import Graph
import igraph as ig
```

```python
n = 13
G = [i for i in range(n)]
E = list(combinations(G[:n//3], 2)) + list(combinations(G[n//3:n//3*2], 2)) + list(combinations(G[n//3*2:], 2))
print(E)
graph = Graph(n,E)
graph.add_edges([(0,n//3), (n//3 + 1, n//3*2), (n//3*2 + 1,1), (1,n//3)])
dend = graph.community_edge_betweenness(clusters=1)
print(graph.community_edge_betweenness(clusters=1))
ig.plot(graph)
```
